```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("insurance.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data)
#get_dummies gets the data in numerical form
data.head()
```

```python
data.drop(['sex_female','smoker_no'],axis = 1, inplace = True)
#inplace - dataframe is modified in place
```

```python
#removed one of the columns to remove redundancy in information
#sex female and sex male show same information
data.rename(columns={'sex_male':'gender','smoker_yes':'smoker'},inplace = True)
```

```python
data.head()
```

```python
# divide dataset into dependent and independent variables
#get the independent variable
X = data.drop('charges',axis = 1)
X.head()
```

```python
#get the dependent variable
Y = pd.DataFrame(data['charges'])
Y.head()
```

```python
#divide the data into training and testing data
from sklearn.model_selection import train_test_split

#80% to training and 20% to testing 
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 2)
#random_state - measure the same training or testing data is generated
#it could generate different training and testing datasets and not sequential
#you can give any value not only 2
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
#simple linear regression
from sklearn.linear_model import LinearRegression
simple_linreg = LinearRegression()
```

```python
#Fit or train the model with training data
#passing one variable
x-smoker
simple_linreg.fit(pd.DataFrame(X_train.smoker), Y_train)
Y_pred = simple_linreg.predict(pd.DataFrame(X_test.smoker))
```

```python
Y_pred
```

```python
print(simple_linreg.score(pd.DataFrame(X_test.smoker),Y_test))
```

```python
#multi linear regression

linreg = LinearRegression()
linreg.fit(X_train,Y_train)
Y_pred_new = linreg.predict(X_test)
```

```python
Y_pred_new
```

```python
print(linreg.score(X_test,Y_test))
#the coefficient of linear regression is the score
#best score is 1
#training is good but testing is bad - overfitting
#check accuracy for both training and testing
```
